```python
class Electronic:
  def __init__(self,name):
    self.name=name
    self.status="OFF"

  def turn_on(self):
    self.status="ON"

  def turn_off(self):
    self.status="OFF"

  def display_status(self):
    print(f"{self.name} is {self.status}")

name=input("Enter the name of the electronic device: ")
action=input("Enter the action to be performed (on/off): ")
device=Electronic(name)
if action=="on":
  device.turn_on()
elif action=="off":
  device.turn_off()
device.display_status()
```

```python
class Employee:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department

  def display(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.department}")

en=input("Employee Name:")
ei=input("Employee ID:")
ed=input("Department:")
employee=Employee(en,ei,ed)
employee.display()
```

```python
class Contact:
  def __init__(self,name,no):
    self.name=name
    self.no=no

  def display(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Number: {self.no}")


name=input("Contact Name:")
no=input("Phone Number:")
contact=Contact(name,no)
contact.display()
```

```python
class Product:
  def __init__(self,pn,p):
    self.pn=pn
    self.p=p

  def display(self):
    print(f"Product:{pn}")
    print(f"Price:₹{p}")

pn=input("Product Name:")
p=input("Price:")
product=Product(pn,p)
product.display()
```

```python
class Rating:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating

  def display(self):
    print(f"Name: {self.name}")
    print(f"Rating: {self.rating}/5")


name=input("Movie name:")
rating=input("Rating:")
rating=Rating(name,rating)
rating.display()
```

```python
class Delivery:
  def __init__(self,customer,address):
    self.customer=customer
    self.address=address

  def display(self):
    print("Delivery Details")
    print(f"Customer: {self.customer}")
    print(f"Address: {self.address}")

customer=input("Customer Name:")
address=input("Address:")
delivery=Delivery(customer,address)
delivery.display()
```
